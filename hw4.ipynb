```python
import numpy as np
import matplotlib.pyplot as plt 
import scipy.stats as stats
%matplotlib inline
```

Problem 0

```python
t = np.linspace(0,np.pi*100, int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
plt.scatter(t,bkgd)
plt.xlabel('time')
plt.ylabel('background')
plt.title('background data vs. time')
plt.show()
```

Problem 1

1a) Make a scatter plot (not a line plot) of the first 1000 data points.

```python
x = t[:1000]
y = bkgd[:1000]
```

```python
plt.scatter(y, x)
plt.xlabel('time')
plt.ylabel('background')
plt.title('background data vs. time')
plt.show()
```

1b) Generalize your code so you can make a plot of any X contiguous points (e.g. 2000 points starting at point 500), and produce an example plot of a set of data somethere in the middle of your array.

```python
def plotContiguous(array, start, end):
    return array[start: end]
```

```python
x = t
y = bkgd
x = plotContiguous(x, 500, 2000)
y = plotContiguous(y, 500, 2000)
plt.scatter(x, y)
plt.title("Background(500 to 2000) vs Time")
plt.xlabel('time')
plt.ylabel('background')
plt.show()
```

1c) Sometimes you want to sample the data, such as plotting every 100th point. Make a plot of the full data range, but only every 100th point.

```python
x = t[::100]
y = bkgd[::100]
plt.scatter(x, y)
plt.title("Background(steps of 100) vs Time")
plt.xlabel('time')
plt.ylabel('background')
plt.show()
```

Problem 2

2a) Make a 2d histogram plot. Read about options, and play with binsize so it is useful (too large a bin and everything washes out, too small and the sample noise dominates). Make sure you use a colorbar.

```python
x = t
y = bkgd
plt.hist2d(x, y, bins = 100) 
plt.title("density plot of background data vs time")
plt.ylabel("background")
plt.xlabel("time")
plt.colorbar()
plt.show()
```

2b) Clearly explain what is being plotted in your plot (e.g. what does each color mean?)

The colorbar represents the number of points coresponding to a particular value of background at a particular time. Yellow corresponds to the maximum number of points and purple corresponds to the minimum number of points.

Problem 3

Make an new time variable that folds at a period of  2π  (the modulo operator is your friend, variably expressed as mod(), %, or rem()). Then make a scatter plot of all your data, but now folded. (Reminder, for more data you'd never actually want to plot all the points.)

```python
t_fold = t % (2*np.pi)
N_fold = 10+np.sin(t_fold)
bkgd_fold = stats.norm.rvs(size= int(1e5))*np.sqrt(N_fold)+N_fold
```

```python
plt.scatter(t_fold[:1000], bkgd_fold[:1000])
plt.title('background vs time for a periodic function')
plt.xlabel('time')
plt.ylabel('background')
plt.show()
```

b) Make a 2D histogram plot of your folded data.

```python
fig,ax = plt.subplots()
plt.hist2d(t_fold, bkgd_fold, bins=100)
plt.xlabel("time")
plt.ylabel("background")
plt.title("density plot for background vs time for a periodic function")
```
