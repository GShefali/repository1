```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import h5py
import pickle
```

```python
infile_qcd = open("qcd_100000_pt_250_500.pkl",'rb')
qcd_data = pickle.load(infile_qcd)
infile_higgs = open("higgs_100000_pt_250_500.pkl", 'rb')
higgs_data = pickle.load(infile_higgs)
```

```python
qcd_data.keys()
```

```python
higgs_data.keys()
```

```python
plt.rcParams["figure.figsize"] = (10,10)

def plot_feature(feature):
    
    
    # plot the background
    plt.subplot(2, 1, 1)
    plt.hist(qcd_data[f'{feature}'], bins= 100)
    plt.title('QCD Background')
    plt.xlabel(f'{feature}')
    plt.ylabel('Event counts')

    # plot the signal
    plt.subplot(2, 1, 2)
    plt.hist(higgs_data[f'{feature}'], bins= 100)
    plt.title('Higgs Boson Signal')
    
    plt.xlabel(f' {feature}')
    plt.ylabel('Event counts')
    
    plt.show()
```

```python
plot_feature('pt')
```

```python
plot_feature('eta')
```

```python
plot_feature('phi')
plot_feature('mass')
plot_feature('ee2')
plot_feature('ee3')
plot_feature('d2')
plot_feature('angularity')
```

```python
plot_feature('t1')
plot_feature('t2')
plot_feature('t3')
plot_feature('t21')
plot_feature('t32')
plot_feature('KtDeltaR')
```
