#### B) Show that after 5 days, the summed probability distribution is still a Poisson distribution. Explain why this makes sense from a mathematical and conceptual point of view.

```python
v2 = stats.poisson.pmf(np.arange(1,10), 5)
v = v2
for x in range(4) :
    v2 = np.convolve(v, v2)
plt.rcParams["figure.figsize"] = (7,5)
fig, ax = plt.subplots(1, 1, sharex = True, sharey = True)
ax.stairs(v2, np.arange(5,47),fill = True)
plt.title("probability of cosmic detection for 5 days on log scale")
ax.set_xlabel("number of cosmic rays")
ax.set_ylabel("probability")
ax.plot(np.linspace(0,60,1000),stats.norm.pdf(np.linspace(0,60,1000),loc = 25, scale = 7),linewidth = 2,alpha = 0.7, label='Gaussian distribution')
ax.set_yscale("log")
plt.legend()
plt.show()
```

Even though from the plot looks like it is approaching a Gaussian distribution from the plot with the log y axis we can see that the distribution is not gaussian and it is slightlty skewed to the left.The distribution is discrete unlike gaussian summing up discrete values does not give continuous values. The distribution is still poisson:

Explanation:
Mathematical : The convolution of independent Poisson distributions is still a Poisson distribution.As the value of mu adds up
mu = mu1 + mu2+ mu3 + m4+...+mun This can be verified using equations.

Conceptual :The distribution is discrete unlike gaussian summing up discrete values does not give continuous values.

#### D) Pick some number of days N, and assume you saw Y*N gamma rays from your source. Calculate the 'sigma' of your observation. [In reality the number of gamma-rays seen from a source will also fluctuate, but we're going to ignore that complication for a couple of labs.]

For 5 days we assume we saw 30 gamma rays.To get the sigma of the observation we get the probability of the sum of detections from the background is 30 or more then we use norm.ppf to convert this probability into significance value 

```python
v2 = stats.poisson.pmf(np.arange(1,10), 5)
v = stats.poisson.pmf(np.arange(1,10), 5)
for x in range(4) :
    v2 = np.convolve(v, v2)
p = v2[30:].sum()    
stats.norm.ppf(p)
```

The sigma value for the observation is 2.3888792275010116
