```python
v2 = stats.poisson.pmf(np.arange(1,10), 5)
for x in range(2) :
    v2 = np.convolve(v2, v2)
plt.rcParams["figure.figsize"] = (7,5)
fig, ax = plt.subplots(1, 1, sharex = True, sharey = True)
ax.stairs( v2,np.arange(10,44),fill = True)

ax.set_xlabel("sum of the number of cosmic rays for 2 days")
ax.set_ylabel("probability")
#ax.plot(np.linspace(10,60,1000),stats.norm.pdf(np.linspace(10,60,1000),loc = 35.5, scale = 5),linewidth = 8,alpha = 0.7)
#ax.set_yscale("log")
plt.show()    
```

Problem 2
Pick a skewed continuous distribtuion, such as a Rayleigh, that describes your background over some observing interval.

A) Show how the distribution changes as you average over more observing intervals.

1 observation

```python
x = np.arange(1,10)
d = stats.rayleigh.pdf(x)
plt.rcParams["figure.figsize"] = (7,5)
fig, ax = plt.subplots(1, 1)
ax.stairs(d,np.arange(1,11),fill = True)
plt.show()
```

2 observations:

```python
x = np.arange(1,10)
d = stats.rayleigh.pdf(x)
d = np.convolve(d, d)
plt.rcParams["figure.figsize"] = (7,5)
fig, ax = plt.subplots(1, 1)
ax.stairs(d,np.arange(2,20)/2, fill = True)
plt.show()
```
